```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df1 = pd.read_csv('athlete_events.csv')
df1.head()
```

```python
df1.shape
```

```python
df2 = pd.read_csv('noc_regions.csv')
df2.head()
```

```python
df2.shape
```

```python
df = df1.merge(df2, how = 'left', on = 'NOC')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isna()
```

```python
df.isnull().sum()
```

```python
df.rename(columns={'region':'Region','notes':'Notes'},inplace = True)
df.head()
```

```python
top_10_countries = df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
plt.figure(figsize=(12,6))
plt.title('Participation by the countries in the olympics')
sb.barplot(x = top_10_countries.index,y = top_10_countries,palette = 'Set1')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title('Age of the participants in the olympics')
plt.hist(df.Age,bins = np.arange(10,80,2),color = 'purple',edgecolor = 'white')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.show()
```

```python
women_in_olympics = df[(df.Sex == 'F') & (df.Season == 'Summer')]
```

```python
sb.set(style = 'whitegrid')
plt.figure(figsize=(20,10))
sb.countplot(x='Year',data = women_in_olympics,palette='Spectral')
plt.title('Women in summer olympics')
plt.show()
```

```python
participants = women_in_olympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(12,6))
participants.loc[:,'F'].plot()
plt.title('Plotting participation of women in summer olympics')
plt.show()
```

```python
gold_medals = df[(df.Medal == 'Gold')]
gold_medals.head()
```

```python
gold_medals = gold_medals[np.isfinite(gold_medals['Age'])]
```

```python
gold_medals['ID'][gold_medals['Age'] > 60].count()
```

```python
sport = gold_medals['Sport'][gold_medals['Age']>60]
sport
```

```python
plt.figure(figsize=(12,6))
plt.tight_layout()
sb.countplot(sport)
plt.title('Sports in which athletes over 60 years won gold medal')
plt.show()
```

```python
gold_medals.Region.value_counts().reset_index(name = 'Medal').head()
```

```python
total_gold_medals = gold_medals.Region.value_counts().reset_index(name = 'Medal').head(5)
#plt.figure(figsize=(20,20))
p = sb.catplot(x = 'index',y = 'Medal',data = total_gold_medals,
               height = 5,kind = 'bar',palette = 'rocket')
p.despine(left=True)
p.set_xlabels('Top 5 countries')
p.set_ylabels('Number of medals')
plt.title('Gold Medals per country')
plt.show()
```

```python
max_year = df.Year.max()
print(max_year)

team_names = df[(df.Year == max_year) & (df.Medal == 'Gold')].Team
team_names.value_counts().head(10)
```

```python
sb.barplot(x = team_names.value_counts().head(20),y=team_names.value_counts().head(20).index)
plt.ylabel(None);
plt.xlabel('Country wise Medals for the year 2016')
plt.figure(figsize=(12,12))
plt.show()
```
